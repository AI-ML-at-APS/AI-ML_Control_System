```python
# Motor vkb 4
translations = np.linspace(-0.15, 0.15, 6)
fig, axes = plt.subplots(2, 3, figsize=[15, 10])
initial_value = focusing_system.get_vkb_motor_4_translation()
opt_trial = opt_common.OptimizationCommon(focusing_system, motor_types='vkb_4', random_seed=DEFAULT_RANDOM_SEED)
for ix, trans in enumerate(translations):
    #focusing_system, _ = opt_common.reinitialize(input_beam_path=input_beam_path)
    opt_trial.setInitialMotorPositions(trans, movement='relative')
    trans_beam = opt_trial.getBeam()
    hist, dw = get_shadow_beam_spatial_distribution(trans_beam)
    h_centroid = dw.get_parameter('h_centroid')
    v_centroid = dw.get_parameter('v_centroid')
    centroid_distance = (h_centroid ** 2 + v_centroid ** 2) ** 0.5
    peak_intensity = dw.get_parameter('peak_intensity')
    axes.flat[ix].pcolormesh(hist.vv, hist.hh, hist.data_2D)
    axes.flat[ix].set_title(f'T {trans:3.2f} P {peak_intensity:3.2f} C {centroid_distance:3.2f}')
    
    opt_trial.setInitialMotorPositions(initial_value)
plt.suptitle('VKB_4')
plt.tight_layout()
plt.show()
```
